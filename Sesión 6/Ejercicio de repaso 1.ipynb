```python
import pandas as pd
import matplotlib.pyplot as plt

asteroid_df = pd.read_json('https://github.com/Marcel-Jan/Fun-with-Python/blob/master/mpcorb_10kasteroids.json?raw=true')
plt.scatter(asteroid_df.a, asteroid_df.e, color='k', marker='.', s=1)
plt.xlim(2, 3.5)
plt.ylim(0, .8)
plt.xlabel('Semimajor axis (AU)')
plt.ylabel('Eccentricity')
plt.title('First 10K asteroids discovered')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import style
style.use("seaborn-paper")


asteroid_df = pd.read_json('https://github.com/Marcel-Jan/Fun-with-Python/blob/master/mpcorb_10kasteroids.json?raw=true')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(asteroid_df.a, asteroid_df.i, asteroid_df.e, color='k', marker='.')

ax.set_xlim(2, 3.5)
ax.set_ylim(0, 20)
ax.set_zlim(0, 0.5)

ax.set_xlabel('Semimajor axis')
ax.set_ylabel('Inclination')
ax.set_zlabel('Eccentricity')
ax.set_title('First 10,000 asteroids discovered')

plt.show()
```
