# Histogramas y densidades

- Muchas veces el primer paso para entender un dataset es un histograma. Anteriormente vimos un histograma sencillo. Veamos algunos ejemplos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

data = np.random.randn(1000)

plt.hist(data);
```

- Personalicemos nuestro histograma

```python
plt.hist(data, bins=30, density=True, alpha=0.5,
         histtype='stepfilled', color='steelblue',
         edgecolor='red');
```

También podemos graficar diferentes histogramas y hacer uso de la propiedad de transparencia cuando comparamos diferentes datos

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

También podemos mostrar solamente los puntos en un bin dado con np.histogram()

```python
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
```

# Histogramas de dos dimensiones y binnings

- Podemos crear histogramas de dos dimensiones separando nuestros puntos en diferentes bins
- Primero creamos nuestros datos de prueba

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

Binings hexagonales

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```
