# Matplotlib

Ahora echaremos un vistazo en profundidad al paquete Matplotlib para su visualización en Python. Matplotlib es una biblioteca de visualización de datos multiplataforma construida sobre matrices NumPy y diseñada para funcionar con la pila SciPy más amplia. Fue concebido por John Hunter en 2002, originalmente como un parche para IPython para permitir el trazado interactivo de estilo MATLAB a través de gnuplot desde la línea de comandos de IPython. El creador de IPython, Fernando Pérez, estaba en ese momento luchando por terminar su doctorado y hacerle saber a John que no tendría tiempo de revisar el parche durante varios meses. John tomó esto como una señal para partir por su cuenta, y nació el paquete Matplotlib, con la versión 0.1 lanzada en 2003. Recibió un impulso temprano cuando fue adoptado como el paquete de trazado elegido por el Instituto de Ciencias del Telescopio Espacial (el personas detrás del telescopio Hubble), que apoyó financieramente el desarrollo de Matplotlib y expandió enormemente sus capacidades.

Una de las características más importantes de Matplotlib es su capacidad para funcionar bien con muchos sistemas operativos y backends gráficos. Matplotlib admite docenas de backends y tipos de salida, lo que significa que puede contar con que funcionará independientemente del sistema operativo que esté utilizando o del formato de salida que desee. Este enfoque multiplataforma, de todo para todos, ha sido una de las grandes fortalezas de Matplotlib. Ha generado una gran base de usuarios, lo que a su vez ha generado una base de desarrolladores activa y las poderosas herramientas y la ubicuidad de Matplotlib dentro del mundo científico de Python.

**Consejos generales de Matplotlib**

Antes de profundizar en los detalles de la creación de visualizaciones con Matplotlib, hay algunas cosas útiles que debe saber sobre el uso del paquete.

Importando Matplotlib
Así como usamos la abreviatura np para NumPy y la abreviatura pd para Pandas, usaremos algunas abreviaturas estándar para las importaciones de Matplotlib:

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

    Configuración de estilos
Usaremos la directiva plt.style para elegir estilos estéticos apropiados para nuestras figuras. Aquí estableceremos el estilo clásico, que asegura que las parcelas que creamos utilicen el estilo clásico de Matplotlib:

```python
plt.style.use('classic')
```

```python
import numpy as np
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

**Etiquetado de graficas** El etiquetado de los gráficos, es decir títulos, etiquetas de ejes y leyendas simples es un detalle muy importante.

Los títulos y las etiquetas de los ejes son las etiquetas más simples de este tipo; existen métodos que se pueden usar para configurarlas rápidamente:

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

La posición, el tamaño y el estilo de estas etiquetas se pueden ajustar utilizando argumentos opcionales para la función. Para obtener más información, consulte la documentación de Matplotlib y las cadenas de documentación de cada una de estas funciones.

Cuando se muestran varias líneas dentro de un solo eje, puede resultar útil crear una leyenda de trazado que etiquete cada tipo de línea. Una vez más, Matplotlib tiene una forma incorporada de crear rápidamente dicha leyenda. Se hace a través del método plt.legend () (lo adivinó). Aunque hay varias formas válidas de usar esto, me resulta más fácil especificar la etiqueta de cada línea usando la palabra clave label de la función de trazado:

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

# Tipos de graficos en matplotlib

**Lineal**

Quizás el más simple de todos los gráficos es la visualización de una sola función $ y = f (x) $. Aquí daremos un primer vistazo a la creación de una trama simple de este tipo. Para todas las gráficas de Matplotlib, comenzamos creando una figura y un eje. En su forma más simple, se pueden crear una figura y ejes de la siguiente manera:

```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
```

Agregamos la información que queramos graficar:

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x));
```

Si queremos crear una sola figura con varias líneas, simplemente podemos llamar a la función de trazado varias veces:

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

La función plt.plot () toma argumentos adicionales que se pueden usar para especificarlos. Para ajustar el color, puede utilizar la palabra clave color, que acepta un argumento de cadena que representa prácticamente cualquier color imaginable. El color se puede especificar de varias formas:

```python
plt.plot(x, np.sin(x - 0), color='blue')        # nombre del color
plt.plot(x, np.sin(x - 1), color='g')           # codigo del color
plt.plot(x, np.sin(x - 2), color='0.75')        # escala de grises
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex  (RRGGBB de 00 a FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, valores 0 a 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # colores soportados por HTML
```

Si no se especifica ningún color, Matplotlib recorrerá automáticamente un conjunto de colores predeterminados para varias líneas.

De manera similar, el estilo de línea se puede ajustar usando la palabra clave linestyle:

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# Tambien se puede usar por codigo:
plt.plot(x, x + 4, linestyle='-')  # solido
plt.plot(x, x + 5, linestyle='--') # guión
plt.plot(x, x + 6, linestyle='-.') # guión punto
plt.plot(x, x + 7, linestyle=':');  # punto
```

**Scatter plot**

Otro tipo de diagrama de uso común es el diagrama de dispersión simple, un primo cercano del diagrama de líneas. En lugar de que los puntos estén unidos por segmentos de línea, aquí los puntos se representan individualmente con un punto, círculo u otra forma. Comenzaremos configurando el cuaderno para graficar e importar las funciones que usaremos:

En la sección anterior, analizamos plt.plot / ax.plot para producir gráficos de líneas. Resulta que esta misma función también puede producir gráficos de dispersión:

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
```

El tercer argumento de la llamada a la función es un carácter que representa el tipo de símbolo utilizado para el trazado. Así como puede especificar opciones como '-', '-' para controlar el estilo de línea, el estilo de marcador tiene su propio conjunto de códigos de cadena cortos. La lista completa de símbolos disponibles se puede ver en la documentación de plt.plot o en la documentación en línea de Matplotlib. La mayoría de las posibilidades son bastante intuitivas y mostraremos algunas de las más comunes aquí:

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

Para obtener aún más posibilidades, estos códigos de caracteres se pueden usar junto con códigos de línea y color para trazar puntos junto con una línea que los conecte:

```python
plt.plot(x, y, '-v');
```

Gráficos de dispersión con dispersión plt.
Un segundo método más poderoso para crear diagramas de dispersión es la función plt.scatter, que se puede usar de manera muy similar a la función plt.plot:

```python
plt.scatter(x, y, marker='o');
```

La principal diferencia de plt.scatter de plt.plot es que se puede utilizar para crear diagramas de dispersión donde las propiedades de cada punto individual (tamaño, color de la cara, color del borde, etc.) se pueden controlar individualmente o asignar a los datos.

Demostremos esto creando un diagrama de dispersión aleatorio con puntos de muchos colores y tamaños. Para ver mejor los resultados superpuestos, también usaremos la palabra clave alfa para ajustar el nivel de transparencia:

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
```

Aparte de las diferentes funciones disponibles en plt.plot y plt.scatter, ¿por qué podría optar por utilizar una sobre la otra? Si bien no importa tanto para pequeñas cantidades de datos, dado que los conjuntos de datos superan unos pocos miles de puntos, plt.plot puede ser notablemente más eficiente que plt.scatter.

# Visualizar Errores

Para cualquier medición científica, la contabilidad precisa de los errores es casi tan importante, si no más importante, que el informe exacto del número en sí. Por ejemplo, imagine que estoy usando algunas observaciones astrofísicas para estimar la Constante de Hubble, la medida local de la tasa de expansión del Universo. Sé que la literatura actual sugiere un valor de alrededor de 71 (km / s) / Mpc, y mido un valor de 74 (km / s) / Mpc con mi método. ¿Los valores son consistentes? La única respuesta correcta, dada esta información, es la siguiente: no hay forma de saberlo.

Suponga que aumento esta información con incertidumbres reportadas: la literatura actual sugiere un valor de alrededor de 71 $ \ pm $ 2.5 (km / s) / Mpc, y mi método ha medido un valor de 74 $ \ pm $ 5 (km / s) / Mpc. Ahora bien, ¿los valores son consistentes? Esa es una pregunta que puede responderse cuantitativamente.

En la visualización de datos y resultados, mostrar estos errores de manera efectiva puede hacer que un gráfico transmita información mucho más completa.

**Barras de error básicas**
Se puede crear una barra de error básica con una sola llamada a la función Matplotlib:

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k');
```

Aquí, el fmt es un código de formato que controla la apariencia de líneas y puntos, y tiene la misma sintaxis que la abreviatura utilizada en plt.plot, descrita en Gráficos de líneas simples y Gráficos de dispersión simples.

Además de estas opciones básicas, la función de barra de errores tiene muchas opciones para ajustar las salidas. Con estas opciones adicionales, puede personalizar fácilmente la estética de su gráfico de barra de error. A menudo me resulta útil, especialmente en parcelas abarrotadas, hacer que las barras de error sean más claras que los puntos en sí:

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0);
```

Además de estas opciones, también puede especificar barras de error horizontales (xerr), barras de error unilaterales y muchas otras variantes. Para obtener más información sobre las opciones disponibles, consulte la cadena de documentos de plt.errorbar.
