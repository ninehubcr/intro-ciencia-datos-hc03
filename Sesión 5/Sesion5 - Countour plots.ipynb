# Plots countour y de densidad

- Muchas veces es útil visualizar datos tri dimensionales en dos dimensiones utilizando regiones de color equivalentes a un valor.
- Para esto podemos utilizar plt.countour, plt.countourf, y plt.imshow. Veamos un ejemplo

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

Primero definamos una función para utilizar como nuestra tercera dimensión

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

Definamos x e y

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

Llamemos nuestra función countour
- Por defecto los valores negativos se van a representar con líneas punteadas, y los positivos con líneas continuas.
- También podemos definir colores para nuestras líneas
- Definimos nuestros colores como 20 intervalos espaciados igualmente

```python
plt.contour(X, Y, Z, colors='black');
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

```python
plt.contour(X, Y, Z, 20, cmap='inferno');
```

```python
plt.contour(X, Y, Z, 20, cmap='viridis');
```

Agreguemos una barra de color

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

```python
plt.contourf(X, Y, Z, 20, cmap='inferno')
plt.colorbar();
```

```python
plt.contourf(X, Y, Z, 20, cmap='viridis')
plt.colorbar();
```

Una forma de mostrar etiquetas sobre nuestros datos es mostrar una imagen con una transparencia definida por encima de nuestros datos graficados, veamos un ejemplo.

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();
```
